```python
#combining heart associators together to form label column
df["event"] = (df["Ever told had congestive heart failure"] == "Yes") | (df["Ever told you had coronary heart disease"] == "Yes") | (df["Ever told you had angina/angina pectoris"] == "Yes") | (df["Ever told you had heart attack"] == "Yes") | (df["Ever told you had a stroke"] == "Yes")
df.head()
```

```python
categorical_val = []
continous_val = []
for column in df.columns:
    print('==============================')
    print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
from sklearn.preprocessing import StandardScaler

s_sc = StandardScaler()
col_to_scale = ['Age in years at screening ', 'Ratio of family income to poverty', '60 sec. pulse (30 sec. pulse * 2)',
                'Weight (kg)','How often drink alcohol over past 12 mos', 'Sodium (mmol/L)', 'Cholesterol, refrigerated serum (mg/dL)',
               'avg_systolic_blood_pres', 'avg_diastolic_blood_pres']
dataset[col_to_scale] = s_sc.fit_transform(dataset[col_to_scale])
```

```python
dataset.head()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print("Classification Report:", end='')
        print(f"\tPrecision Score: {precision_score(y_train, pred) * 100:.2f}%")
        print(f"\t\t\tRecall Score: {recall_score(y_train, pred) * 100:.2f}%")
        print(f"\t\t\tF1 score: {f1_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print("Classification Report:", end='')
        print(f"\tPrecision Score: {precision_score(y_test, pred) * 100:.2f}%")
        print(f"\t\t\tRecall Score: {recall_score(y_test, pred) * 100:.2f}%")
        print(f"\t\t\tF1 score: {f1_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
       
```

```python
from sklearn.model_selection import train_test_split

X = dataset.drop('event', axis=1)
y = dataset.event

#splitting into 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(class_weight='balanced')
log_reg.fit(X_train, y_train) # fitting classifier to training set
```

```python
print_score(log_reg, X_train, y_train, X_test, y_test, train=True)
print_score(log_reg, X_train, y_train, X_test, y_test, train=False)
```

```python
test_score = accuracy_score(y_test, log_reg.predict(X_test)) * 100
train_score = accuracy_score(y_train, log_reg.predict(X_train)) * 100

#appending results into a dataframe
results_df = pd.DataFrame(data=[["Logistic Regression", train_score, test_score]], 
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df
```

```python
from sklearn.tree import DecisionTreeClassifier


tree = DecisionTreeClassifier(class_weight='balanced')
tree.fit(X_train, y_train)

print_score(tree, X_train, y_train, X_test, y_test, train=True)
print_score(tree, X_train, y_train, X_test, y_test, train=False)
```

```python
test_score = accuracy_score(y_test, tree.predict(X_test)) * 100
train_score = accuracy_score(y_train, tree.predict(X_train)) * 100

results_df_2 = pd.DataFrame(data=[["Decision Tree Classifier", train_score, test_score]], 
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df = results_df.append(results_df_2, ignore_index=True)
results_df
```

```python
from sklearn.svm import SVC


svm_model = SVC(class_weight='balanced')
svm_model.fit(X_train, y_train)
```

```python
print_score(svm_model, X_train, y_train, X_test, y_test, train=True)
print_score(svm_model, X_train, y_train, X_test, y_test, train=False)
```

```python
test_score = accuracy_score(y_test, svm_model.predict(X_test)) * 100
train_score = accuracy_score(y_train, svm_model.predict(X_train)) * 100

results_df_2 = pd.DataFrame(data=[["Support Vector Machine", train_score, test_score]], 
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df = results_df.append(results_df_2, ignore_index=True)
results_df
```

# Ensemble Method

```python
# from hands on machine learning with Scikit-Learn & TensorFlow
from sklearn.ensemble import VotingClassifier
log_clf = LogisticRegression(random_state=42, class_weight='balanced')
dt_clf = DecisionTreeClassifier(random_state=42, class_weight='balanced')
svm_clf = SVC(random_state=42, class_weight='balanced')
voting_clf = VotingClassifier(estimators=[('lr', log_clf), ('dt', dt_clf), ('svc', svm_clf)],
voting='hard')
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
for clf in (log_clf, dt_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))#model accuracy
```

## classifiers without any tuning

```python
# Instantiate the classfiers and make a list
classifiers = [LogisticRegression(class_weight='balanced'),   
               DecisionTreeClassifier(class_weight='balanced'),
               SVC(class_weight='balanced', probability=True)]

# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])

# Train the models and record the results
for cls in classifiers:
    model = cls.fit(X_train, y_train)
    yproba = model.predict_proba(X_test)[::,1]
    
    fpr, tpr, _ = roc_curve(y_test,  yproba)
    auc = roc_auc_score(y_test, yproba)
    
    result_table = result_table.append({'classifiers':cls.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)

# Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)
```

```python
fig = plt.figure(figsize=(8,6))

for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.2f}".format(i, result_table.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("Flase Positive Rate", fontsize=15)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)

plt.title('ROC Curve Analysis for base classifiers', fontweight='bold', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')

plt.show()
```

```python
# Instantiate the classfiers and make a list
tunedclassifiers = [LogisticRegression(C=10, solver='saga', class_weight='balanced'),   
               DecisionTreeClassifier(criterion='gini', 
                              max_depth=19,
                              min_samples_leaf=1, 
                              min_samples_split=2, 
                              splitter='best',
                             class_weight='balanced'),
               SVC(C=0.01, gamma=0.0001, kernel='poly', class_weight='balanced', probability=True)]


# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=['tunedclassifiers', 'fpr','tpr','auc'])

# Train the models and record the results
for cls in tunedclassifiers:
    model = cls.fit(X_train, y_train)
    yproba = model.predict_proba(X_test)[::,1]
    
    fpr, tpr, _ = roc_curve(y_test,  yproba)
    auc = roc_auc_score(y_test, yproba)
    
    result_table = result_table.append({'tunedclassifiers':cls.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)

# Set name of the classifiers as index labels
result_table.set_index('tunedclassifiers', inplace=True)
```
